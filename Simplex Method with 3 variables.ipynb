# Simplex Method with 3 variables 

````python
from scipy.optimize import linprog
````

### #Min z = x1-3x2+2x3
### #subject to
### #3x1-x2+3x3<=7
### #-2x1+4x2<=12
### #-4x1+3x2+8x3<=10
### #x1,x2,x3 >=0

```
# This is formatted as code
```



````python
obj = [1,-3,2]
````

````python
lhs_ineq = [[3,-1,3],
            [-2,4,0],
            [-4,3,8]]
````

````python
rhs_ineq = [7,
            12,
            10]
````

````python
bnd = [(0, float("inf")), #bounds of x
       (0,float("inf")),
       (0,float("inf"))]  #bounds of y
````

````python
opt = linprog(c=obj,A_ub =lhs_ineq, b_ub = rhs_ineq,
              bounds=bnd,method="revised simplex")
opt
````
