```python
Max z= x+2y
subject to
2x+y<=20
-4x+5y<=10
-x+2y>= -2
-x+5y=15
x,y>=0
```

```python
from scipy.optimize import linprog
```

```python
obj = [-1,-2]
```

```python
lhs_ineq = [[2,1],
            [-4,5],
            [1,-2]]
```

```python
rhs_ineq = [20,
            10,
            2]
```

```python
lhs_eq = [[-1,5]]
```

```python
rhs_eq = [15]
```

```python
bnd = [(0, float("inf")), #bounds of x
       (0,float("inf"))]
```

```python
opt = linprog(c=obj,A_ub =lhs_ineq, b_ub = rhs_ineq,A_eq=lhs_eq,b_eq=rhs_eq,
              bounds=bnd,method="revised simplex")
opt
```
