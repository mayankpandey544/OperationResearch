```python
from scipy.optimize import linprog
```

Max z = 3x1 + 2x2
subject to 
x1 + x2 <=4
x1 - x2 <=2
x1,x2 >= 0

```python
obj = [-3,-2]
```

```python
lhs_ineq = [[1,1],
            [1,-1]]
```

```python
rhs_ineq = [4,
            2]
```

```python
bnd = [(0, float("inf")),
       (0,float("inf"))]
```

```python
opt = linprog(c=obj,A_ub =lhs_ineq, b_ub = rhs_ineq,
              bounds=bnd,method="revised simplex")
opt
```
